```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.corpus import words 
#nltk.download("words")
import nltk
```

```python
english_words = set(words.words())
english_words
```

```python
stops = set(stopwords.words('english'))
stops
```

```python
orig_word_df = pd.read_csv('data/unigram_freq.csv')
orig_word_df
```

```python
cleaned_df = orig_word_df[orig_word_df['word'].apply(lambda word: word not in stops)].reset_index(drop = True)
cleaned_df
```

```python
cleaned_df = cleaned_df[cleaned_df['word'].apply(lambda x: x in english_words)].reset_index(drop = True)
cleaned_df
```

```python
cleaned_df.to_csv('data/unigram_freq_clean.csv')
```
